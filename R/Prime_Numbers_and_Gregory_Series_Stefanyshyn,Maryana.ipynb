PROBLEM 1

Write R code to return a vector of prime numbers within range from 0 to 1000.

```r
findprime = 0:1000

for(i in findprime){
  if (i ==0 ){
    next
  } else if (i == 1){
    next
  } else if(i == 2){
    i = 2
  } else if (i%% 2 == 0){
    next
  } else if (i == 3){
    i = 3
  } else if (i %% 3 == 0){
    next
  } else if (i == 5){
    i= 5
  } else if (i %% 5 == 0){
    next
  } else if (i==7){
      i=7
  } else if (i%%7==0){
    next
  }
    primenum=c(i)

    print(primenum)

}

#has to be greater than 1

#find multiples of 2, remove them , mark 2 as prime number

#find multiples of 3, remove them , mark 3 as prime number

##skip 4 & 6 its been canceled out (multiple of 2)

#find multiples of 5, remove them

#find multiples of 7 and remove them

#only go up to number 10
```

We only need to do this process till sqrt(100). why?

Because, all the multiples have been cancelled out when looking at numbers 1 to 10. This means you only do the process untill number 10, which is the sqrt(100)
