```r
#1.21 Load the data set primes (UsingR). This is the set of prime numbers in [1,2003]. How many are there? How many in the range [1,100]? [100,1000]?

library(MASS)
library(HistData)
library(Hmisc)
library(lattice)
library(survival)
library(Formula)
library(ggplot2)
library(UsingR)

data("primes")

print("How many are there?")
length(primes)

print("How many in the range[1,100]?")
length(primes[primes<=100])

print("How many in the range[100,1000]?")
length(primes[primes>=100 & primes<=1000])
```

```r
#1.22 Load the data set primes (UsingR). We wish to find all the twin primes. These are numbers p and p+2, where both are prime

data("primes")

print("Explain what primes [-1] returns")
primes[-1]
print("Answer: it returns all prime numbers except for the first prime number")


print("Explain what primes [-n] returns")
n=length(primes)
primes[-n]
print("Answer: it returns all prime numbers except for the last prime number")


print("How many twin primes are in the data set?")

print("Answer: primes[-1] and primes[-n] give the difference between two prime numbers. Using this, we can count how many variables are equal to 2 in this vector and find how many twin pairs there are ")

sum(primes[-1]-primes[-length(primes)]==2)
```

```r
#1.24 The data set mandms (UsingR) contains the targeted color distribution in a bag of M&Ms as percentages for varies types of packaging. Answer these questions

#color distribution in bag of M&Ms

data("mandms")

print("Which packaging is missing one of the six colors?")
names(which(rowSums(mandms==0)==1))

print("Which types of packaging have equal distribution of colors?")
names(which(rowSums(mandms==rowMeans(mandms))==6))

print("Which packaging has a single color that is more likely than all the others? What color is this?")

names(which(rowSums(mandms==max(mandms))==1))
names(which(colSums(mandms==max(mandms))==1))
```

```r
#1.25 The times variable in the data set nym. 2002 (UsingR) contains the time to finish for several participants in the 2002 New York City Marathon. Answer these questions

data("nym.2002")

print("How many times are stored in the data set?")
length(nym.2002$time)

print("What was the fastest time in minutes?")
fastest=min(nym.2002$time)
paste(fastest %/% 60, round(fastest %% 60), sep=":")

print("What was the slowest time in minutes?")
slowest=max(nym.2002$time)
paste(slowest %/% 60, round(slowest %% 60), sep=":")
```
